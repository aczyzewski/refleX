## Fast.ai
### Uwagi:
- fast.ai tworzy zbior walidujacy na podstawie zbioru trenujacego, w obecnej
    implementacji mozna co najwyzej zdecydowac, ktore indeksy z pliku CSV trafiaja
    do zbioru walidacyjnego. W tej sytuacji, zrobilem merga zbioru walidujacego
    i trenujacego (w tej kolejnosci), gdzie pozniej jako parametr zbioru walidacyjnego 
    podaje pierwsze 200 indeksow. Stąd parametr: `validate_train_csv`.
          
- fast.ai wymaga rozgraniczenia na glowny folder z danymi (root) i subfolder 
    z konkretnymi obrazami. Stad rozgraniczenie na `images_path` i `images_folder`.
          
- fast.ai wymaga innego typu pliku CSV w postaci: `<obraz>`, `<anomalia1>`, `<anomalia2>` ... Pliki CSV z przedrostiem `fastai_` są plikami przekształconymi.

# Learner comparison

```python
from defined_metrics import get_predefined_metrics, get_class_specific_metrics
import pandas as pd
import pprint
from joblib import load, dump
```

```python
RESULT_PATH = '../results/learner_comparison/'
classifier_names = ["SVM", "GaussianNB", "RF", "KNN"]
```

```python
for classifier_name in classifier_names:
    params = vector_parameters.copy()
    params['name'] = classifier_name
    params['output_model_name'] = f'{classifier_name}-grid_search.joblib'
    vc = VectorLearner(params)
    vc.train()
    dump(vc, f'{RESULT_PATH}scikit_models/{classifier_name}.joblib')
```

```python
classifiers =  [load(f'{RESULT_PATH}scikit_models/{classifier_name}.joblib') for classifier_name in classifier_names]
    
classifiers += [FastaiLearner(cartesian_parameters), 
                FastaiLearner(polar_parameters_min),
                FastaiLearner(polar_parameters_max), 
                #JointLearner()
               ]
```

```python
general_results_raw = pd.DataFrame(columns=['classifier'] + [metric.__name__ for metric in get_predefined_metrics()])
class_specific_results_raw = pd.DataFrame(columns=['classifier'] + [metric.__name__ for metric in get_class_specific_metrics()])

pp = pprint.PrettyPrinter(indent=4)

for clf in classifiers:
    values = clf.test()
    pp.pprint(values)
    
    values[0]['classifier'] = values[1]['classifier'] = clf.get_name()
    
    general_results_raw = general_results_raw.append(values[0], ignore_index=True)
    class_specific_results_raw = class_specific_results_raw.append(values[1], ignore_index=True)
    
general_results_raw.to_csv(f'{RESULT_PATH}general_results_raw.csv')
class_specific_results_raw.to_csv(f'{RESULT_PATH}class_specific_results_raw.csv')
```

```python
RESULTS_DIRECTORY = '../results/simple_classifiers/'
DATA_DIRECTORY = f'{RESULTS_DIRECTORY}data_separate/'
CSV_FILENAME = 'simple_classifiers.csv'
```

```python
file_paths = glob.glob(f'{DATA_DIRECTORY}*')
```

```python
joined_df = pd.DataFrame()
for file_path in file_paths:
    df = pd.read_csv(file_path, index_col=0)
    df['anomaly'] = file_path[file_path.rfind('/')+1:file_path.rfind('.')]
    joined_df = joined_df.append(df, ignore_index=True)
joined_df = gather(joined_df, 'measure_name', 'measure_value', ['accuracy', 'precision', 'recall', 'recall_macro', 'roc_auc', 'f1'])
joined_df
```

```python
mean_df = joined_df.groupby(['measure_name', 'clf', 'anomaly'])['measure_value'].mean().reset_index()
std_df = joined_df.groupby(['measure_name', 'clf', 'anomaly'])['measure_value'].std().reset_index()

aggregated = mean_df.rename(index=str, columns={'measure_value':'mean'})
aggregated.loc[:,'std'] = std_df.loc[:, 'measure_value'].values
aggregated 
```

```python
aggregated.to_csv(f'{RESULTS_DIRECTORY}{CSV_FILENAME}')
```

```python
p = ggplot(aggregated, aes(x='clf', y='mean', fill='clf')) + geom_point() + scale_fill_manual(paul_tol_colors)
p = p + geom_bar(stat='identity')
p = p + geom_errorbar(aes(ymin='mean-std', ymax='mean+std'))
p = p + facet_grid('anomaly~measure_name') + theme_minimal()
ggsave(p, f'{RESULTS_DIRECTORY}wykresik.svg')
ggsave(p, f'{RESULTS_DIRECTORY}wykresik.png', width=15, height=15)
```
