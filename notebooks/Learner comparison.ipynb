## Fast.ai
### Uwagi:
- fast.ai tworzy zbior walidujacy na podstawie zbioru trenujacego, w obecnej
    implementacji mozna co najwyzej zdecydowac, ktore indeksy z pliku CSV trafiaja
    do zbioru walidacyjnego. W tej sytuacji, zrobilem merga zbioru walidujacego
    i trenujacego (w tej kolejnosci), gdzie pozniej jako parametr zbioru walidacyjnego 
    podaje pierwsze 200 indeksow. Stąd parametr: `validate_train_csv`.
          
- fast.ai wymaga rozgraniczenia na glowny folder z danymi (root) i subfolder 
    z konkretnymi obrazami. Stad rozgraniczenie na `images_path` i `images_folder`.
          
- fast.ai wymaga innego typu pliku CSV w postaci: `<obraz>`, `<anomalia1>`, `<anomalia2>` ... Pliki CSV z przedrostiem `fastai_` są plikami przekształconymi.

```python
joined_df = pd.DataFrame()
for file_path in file_paths:
    df = pd.read_csv(file_path, index_col=0)
    df['anomaly'] = file_path[file_path.rfind('/')+1:file_path.rfind('.')]
    joined_df = joined_df.append(df, ignore_index=True)
joined_df = gather(joined_df, 'measure_name', 'measure_value', ['accuracy', 'precision', 'recall', 'recall_macro', 'roc_auc', 'f1'])
joined_df
```

```python
mean_df = joined_df.groupby(['measure_name', 'clf', 'anomaly'])['measure_value'].mean().reset_index()
std_df = joined_df.groupby(['measure_name', 'clf', 'anomaly'])['measure_value'].std().reset_index()

aggregated = mean_df.rename(index=str, columns={'measure_value':'mean'})
aggregated.loc[:,'std'] = std_df.loc[:, 'measure_value'].values
aggregated 
```

```python
aggregated.to_csv(f'{RESULTS_DIRECTORY}{CSV_FILENAME}')
```
